# Living Switzerland - Hydroperiod

---
--> Once you have the water masks, which is the major code that I tend to keep by me, the hydroperiod is just a) the interpolation between the two dates, and b) then the accumulation (counting) of days that a pixel is inundated or not.Â  
For a: 
case I: inundated on one date (day1) and inundated the continuous next date (day2) = all the days from day1 to day2 is the pixel inundated.
case II: inundated on one date (day1) and non-inundated the continuous next date (day2) (or vice versa) = half the days from day1 to day2 is the pixel inundated.
case III: non inundated on one date (day1) and non inundated the continuous next date (day2) = all the days from day1 to day2 is the pixel non inundated.
For b:
add the cases per pair of dates' assessments (minus the common dates among pairs - to account for duplicates).

It is simple, evidence - as possible - based, and becomes more accurate to the real situation the more frequent the acquisitions become.
